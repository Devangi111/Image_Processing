```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

## Thresholding - It's process of setting image's pixels value either 0 or 255 (maximum). It can be performed on grayscale or color images.

### Various techniques are there to perform Thresholding
1.   Binary Thresholding - It will convert pixel value to zero below threshold and pixel value to 255 (max) above threshold. 
2.   Inverted Thresholding - It will convert pixel value to 255 (max) below threshold and pixel value to 0 below threshold. It's inverted process of Binary Thresholding.
3.   Truncated Thresholding - The pixel values below the threshold are left unaltered, and all other values are set to the threshold value
4.   Threshold To Zero - It is the thresholding process of setting the pixel values below the threshold to zero while the pixel values above the threshold are left unaltered. 
5.   Threshold To Zero Inverse - It is the thresholding process of setting the pixel values above the threshold to zero while the pixel values below the threshold are left unaltered. 

```python
# read a gray scale image
img=cv2.imread('Cute_Baby.jpg',0)
img = cv2.resize(img, (320,225))

# apply various thresholds
val, th1 = cv2.threshold(img, 110, 255, cv2.THRESH_BINARY)
val, th2 = cv2.threshold(img, 110, 255, cv2.THRESH_BINARY_INV)
val, th3 = cv2.threshold(img, 110, 255, cv2.THRESH_TRUNC)
val, th4 = cv2.threshold(img, 110, 255, cv2.THRESH_TOZERO)
val, th5 = cv2.threshold(img, 110, 255, cv2.THRESH_TOZERO_INV)

# create figure
fig = plt.figure(figsize=(15, 7))

# setting values to rows and column variables
rows = 2
columns = 3

# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)
  
# showing image
plt.imshow(img)
plt.axis('off')
plt.title("Original")
  
# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)
  
# showing image
plt.imshow(th1)
plt.axis('off')
plt.title("Binary Thresholding")
  
# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)
  
# showing image
plt.imshow(th2)
plt.axis('off')
plt.title("Inverse Binary Thresholding")
  
# Adds a subplot at the 4th position
fig.add_subplot(rows, columns, 4)
  
# showing image
plt.imshow(th3)
plt.axis('off')
plt.title("Truncated Thresholding")

# Adds a subplot at the 5th position
fig.add_subplot(rows, columns, 5)
  
# showing image
plt.imshow(th4)
plt.axis('off')
plt.title("Threshold To Zero")

# Adds a subplot at the 6th position
fig.add_subplot(rows, columns, 6)
  
# showing image
plt.imshow(th5)
plt.axis('off')
plt.title("Threshold To Zero Inverse")
```

## Rather than dealing with fixed thresholding there is provison of Adaptive Thresholding too. Adaptive thresholding determines the threshold values locally.

### Types of Adaptive Thresholding

1. Mean Adaptive Thresholding - It will calculate arithmetic mean of neighbourhood pixels.
2. Gaussian Adaptive Thresholding - Works on excluding noise technique.

```python
img = cv2.imread('Cute_Baby.jpg',0)
img = cv2.resize(img, (320,225))
# apply various adaptive thresholds
th1 = cv2.adaptiveThreshold(img, 255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY, 7, 4)
th2 = cv2.adaptiveThreshold(img, 255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY, 7, 4)

# create figure
fig = plt.figure(figsize=(15, 7))

# setting values to rows and column variables
rows = 1
columns = 3

# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)
  
# showing image
plt.imshow(img)
plt.axis('off')
plt.title("Original")
  
# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)
  
# showing image
plt.imshow(th1)
plt.axis('off')
plt.title("Mean Adaptive Thresholding")

# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)
  
# showing image
plt.imshow(th2)
plt.axis('off')
plt.title("Gaussian Adaptive Thresholding")
```

## Image smoothing 

### Image smoothing is an important image processing technique that performs blurring and noise filtering in an image. It finds applications in preprocessing and postprocessing of deep learning models.

### Smoothing is peformed by 2D kernel of a specific size on channels of the image.

#### Following are types of smoothing performed
1. Blur - It is the simple filter that homogeneously applies its kernel to calculate the local weighted average.
2. Gaussian Blur - It is the process of removing Gaussian noise from an image. This method takes the image, the kernel size and the standard deviation as arguments.
3. Median Blur - It is known for its salt-and-pepper noise removal. Unwanted small white and black dots on an image is removed with this tool.
4. Bilateral Filtering - It is used when there is a need for both noise filtering and edges retention. This method detects sharp edges and keeps it as such without any blur.

```python
img = cv2.imread('Cute_Baby.jpg', 1)
img = cv2.resize(img, (300,300))
# Apply blur
img_blur = cv2.blur(img,(3,3))

#Apply Gaussian Blur
img_gaussian = cv2.GaussianBlur(img,(5,5),2)
    
#Apply Median Blur
img_median = cv2.medianBlur(img,3)
    
#Apply Bilateral Filtering
img_bilateral = cv2.bilateralFilter(img,7,100,100)

# create figure
fig = plt.figure(figsize=(15, 7))

# setting values to rows and column variables
rows = 2
columns = 3

# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)
  
# showing image
plt.imshow(img)
plt.axis('off')
plt.title("Original")
  
# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)
  
# showing image
plt.imshow(img_blur)
plt.axis('off')
plt.title("Blur")
  
# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)
  
# showing image
plt.imshow(img_gaussian)
plt.axis('off')
plt.title("Gaussian Blur")
  
# Adds a subplot at the 4th position
fig.add_subplot(rows, columns, 4)
  
# showing image
plt.imshow(img_median)
plt.axis('off')
plt.title("Median Blur")

# Adds a subplot at the 5th position
fig.add_subplot(rows, columns, 5)
  
# showing image
plt.imshow(img_bilateral)
plt.axis('off')
plt.title("Bilateral Filtering")
```

## Canny Edge Detection

### Canny Edge Detection is a powerful edge detection algorithm that performs Gaussian noise filtering, Sobel based horizontal and vertical edges detection, non-maximum suppression to remove unwanted edge points and hysteresis thresholding with two limiting thresholds to have thin and strong edges.

Please run below code and by changing the minimum and maximum threshold values, we can improve the results.

```python
img = cv2.imread('Taj_Mahal_India_Wallpaper.jpg', 0)
img = cv2.resize(img, (450,300))

def null(x):
    pass
# create trackbars to control threshold values
cv2.namedWindow('Canny')
cv2.resizeWindow('Canny', (450,300))
cv2.createTrackbar('MIN', 'Canny', 80,255, null)
cv2.createTrackbar('MAX', 'Canny', 120,255, null)
while True:
     # get Trackbar position
     a = cv2.getTrackbarPos('MIN', 'Canny')
     b = cv2.getTrackbarPos('MAX', 'Canny')
     # Canny Edge detection
     # arguments: image, min_val, max_val
     canny = cv2.Canny(img,a,b)
     # display the images
     cv2.imshow('Canny', canny)
     k = cv2.waitKey(1) & 0xFF
     if k == ord('q'):
         break
cv2.destroyAllWindows()
```

## Image Contours

### Image contours are the continuous shape outlines present in an image. OpenCV detects the contours present in an image and collects its coordinates as a list. The collected contours can be drawn over the original image back.

Contours are the fundamental building blocks for object shape detection, motion detection and image segmentation. Contours are collected in a Python list.

```python
img = cv2.imread('cute_baby.jpg', 1)
img = cv2.resize(img, (320,480))

# binary thresholding
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
val,th = cv2.threshold(gray, 127,255,cv2.THRESH_BINARY)
# find contours
contours,_ = cv2.findContours(th, 
                           cv2.RETR_TREE,
                           cv2.CHAIN_APPROX_NONE)
# draw contours on original image
# arguments: image, contours list, index of contour, colour, thickness
cv2.drawContours(img, contours, -1, (0,0,255),1)

# create figure
fig = plt.figure(figsize=(15, 7))

plt.imshow(img)
```
